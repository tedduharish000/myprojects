```
from pyspark.sql import SparkSession
from pyspark.sql.functions import expr
from pyspark.sql.functions import col,isnan,when,count
from pyspark.sql.functions import *
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("RemoveColumnSpaces").getOrCreate()
```

```
df = spark.read.csv('/FileStore/tables/retail_sales_dataset.csv', header = True , inferSchema=True)
# Replace spaces and special characters with underscores
df = df.toDF(*[col.replace(" ", "_") for col in df.columns])
df.display()

```

```
df.printSchema()
```

```
df.describe().display()
```

```
df.select([count(when(col(c).isNull(), c)).alias(c) for c in df.columns]).display()
```

```
df.display()
```

```
df.write.format("delta").mode("overwrite").saveAsTable("my_delta_table")

```

```
%sql
Show tables;
select * from my_delta_table


```

```
from pyspark.sql.functions import to_date, month, sum

monthly_sales = df.groupBy(month("Date").alias("month")).agg(sum("Total_Amount").alias("total_sales"))
monthly_sales.show()
```

```
category_sales = df.groupBy("product_category").agg(sum("Total_Amount").alias("total_sales"))
category_sales.show()
```
